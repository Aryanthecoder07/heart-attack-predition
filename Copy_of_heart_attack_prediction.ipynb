```python
a=pd.read_csv("/kaggle/input/heart-attack/Heart Attack Data Set.csv")
a.head()
```

```python
x=a.iloc[:,:-1]
y=a.iloc[:,-1]
```

```python
from sklearn.feature_selection import SelectKBest,f_regression
sf=SelectKBest(f_regression,k='all')
sf.fit(x,y)
```

```python
d=sf.scores_
d
```

```python
df=[i for i,score in zip(x.columns,d) if score<10]
```

```python
df
```

```python
x1=x.drop(columns=df)
```

```python

from sklearn.preprocessing import StandardScaler
st=StandardScaler()
x1=st.fit_transform(x1)
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x1,y,test_size=0.2)
```

```python
import tensorflow
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam
```

```python
model=Sequential()
model.add(Dense(10,activation='relu',input_dim=9))
model.add(Dense(10,activation='relu'))
model.add(Dense(10,activation='relu'))
model.add(Dense(1,activation='sigmoid'))
```

```python
model.summary()
```

```python
model.compile(optimizer=Adam(learning_rate=0.005),loss='binary_crossentropy',metrics=['accuracy'])
```

```python
model.fit(x_train,y_train,epochs=100)
```

```python
y1=model.predict(x_test)
```

```python
y1
```

```python
y2=np.where(y1>0.5,1,0)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y2)
```
